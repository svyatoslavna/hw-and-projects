## Задание 3: первичный анализ

Выведем статистику для каждого файла по образцу:

```python
def analyse_file(filename): # чтоб поменьше писать, сделала функцию

    # 1. Разбиваем содержание файла на строки
    with open(filename, 'r', encoding='utf-8') as file:
        dataset_lines = [line.strip('\n') for line in file]

    # 2. Считаем статистику (используйте цикл for и вот такие счетчики >)
    total_lines = 0         # общее количество строк
    total_words = 0         # общее количество слов
    total_chars = 0         # общее количество символов
    non_empty_lines = 0     # количество непустых строк (длина такой строки больше 0)
    non_empty_lines_list = list()

    for line in dataset_lines:
        total_lines += 1
        total_words += len(line.split())
        total_chars += len(line)
        if line:
           non_empty_lines += 1
           non_empty_lines_list.append(line)


    # 3. Анализ длин строк
    max_line_length = 0   # максимальная длина строки
    min_line_length = 0   # минимальная длина строки

    for line in non_empty_lines_list:
        max_line_length = len(line) if len(line) > max_line_length else max_line_length

    min_line_length = max_line_length

    for line in non_empty_lines_list:
        min_line_length = len(line) if len(line) < min_line_length else min_line_length

    # Выводим результат на экран для каждого файла в таком формате:
    print(f"=== ИНФОРМАЦИЯ ДЛЯ {filename} ===")
    print(f"Всего строк: {total_lines}")
    print(f"Непустых строк: {non_empty_lines}")
    print(f"Всего слов: {total_words}")
    print(f"Всего символов: {total_chars}")
    print(f"Макс. длина строки: {max_line_length}")
    print(f"Мин. длина строки: {min_line_length}\n")


print("ПЕРВИЧНЫЙ АНАЛИЗ\n")
analyse_file("twitter.txt")
analyse_file("twitter_conv.txt")
analyse_file("movie.txt")
analyse_file("bnc_split.txt")
analyse_file("bnc.txt")
```

## Задание 4: чистка текста

```python
import re

def clean_text(filename):
    cleaned_lines = [] # чистим текст построчно (сохраняем каждую очищенную строку в список)
    all_cleaned_words = [] # также сохраняем все слова после очистки в список

# Ваш код здесь: реализуйте очистку с помощью цикла for и RegEx
# 1. Используйте strip для удаления лишних пробелов из строки
# 2. Очищаем строку: оставляем только буквы, цифры и пробелы
# 3. Убираем лишние пробелы
# 4. Приводим к нижнему регистру
# 5. Собираем все слова после очистки в единый список
#    - не фиксируем слова короче 1 символа   # М: Вы имели в виду "короче 2"?

    # М: тут заново открываю файл, чтобы эта ф-ия работала
    # без предварительного вызова analyse_file():
    with open(filename, "r", encoding="utf-8") as file:
        dataset_lines = [line.strip() for line in file] # strip

        for line in dataset_lines:
            line = re.sub(r"[^\w\s]|_", " ", line) # буквы, цифры, пробелы
            line = re.sub(r"\s+", " ", line) # лишние пробелы
            line = line.lower() # нижний регистр
            if line:
                cleaned_lines.append(line)
                words = line.split()
                # не фиксировала слова короче 2 символов:
                valid_words = [word for word in words if len(word) > 1]
                all_cleaned_words.extend(valid_words)

    # Выводим результат для каждого текста по образцу
    print(f"=== ИНФОРМАЦИЯ ДЛЯ {filename} ===")
    print(f"Пример очищенной строки: {cleaned_lines[0][:100]}...")
    print(f"Всего очищенных слов: {len(all_cleaned_words)}\n")
    return filename, all_cleaned_words

print("ЧИСТКА ТЕКСТА\n")
filename1, all_cleaned_words1 = clean_text("twitter.txt")
filename2, all_cleaned_words2 = clean_text("twitter_conv.txt")
filename3, all_cleaned_words3 = clean_text("movie.txt")
filename4, all_cleaned_words4 = clean_text("bnc_split.txt")
filename5, all_cleaned_words5 = clean_text("bnc.txt")
```

## Задание 5: статистика

Часть 1: получаем статистику распределений длин слов

```python
# Статистика по очищенным данным

def print_statistics(filename, all_cleaned_words):
    # М: данная ф-ия будет работать только после чистки, что, кажется, логично

    # список длин слов и общее количество символов
    total_words = len(all_cleaned_words)
    word_lengths = [len(word) for word in all_cleaned_words]
    total_chars = sum(word_lengths)

    # анализ длин слов
    word_length_count = {}

    for length in word_lengths:
        if length in word_length_count:
            word_length_count[length] += 1 # если длина уже зафиксирована
        else:
            word_length_count[length] = 1 # если длина ещё не зафиксирована

    # результат:
    print(f"=== ИНФОРМАЦИЯ ДЛЯ {filename} ===")
    print(f"Слов после очистки: {total_words}")
    print(f"Символов после очистки: {total_chars}")
    print("Распределение длин слов: ")
    for length in sorted(word_length_count.keys()):
        print(f"  {length} букв: {word_length_count[length]} слов")
    print("\n")


print("СТАТИСТИКА ПО ОЧИЩЕННЫМ ДАННЫМ\n")
print_statistics(filename1, all_cleaned_words1)
print_statistics(filename2, all_cleaned_words2)
print_statistics(filename3, all_cleaned_words3)
print_statistics(filename4, all_cleaned_words4)
print_statistics(filename5, all_cleaned_words5)
```
